```python
db, companies = mc.connectCollection('datlab','companies_seattle')
db, starbucks = mc.connectCollection('datlab', 'starbucks')
db, schools = mc.connectCollection('datlab', 'schools')
db, nightclubs = mc.connectCollection('datlab', 'nightclubs')
db, vegan = mc. connectCollection('datlab', 'vegan_restaurants')
db, kinder = mc.connectCollection('datlab', 'kindergarten')
```

Here I get the location of the facilities around of the top 3 offices ranked in the Geoqueries notebook

OPTIFY

```python
optify = list(companies.find({"name":"Optify"}))
```

```python
coordinates = optify[0]['location']['coordinates']
```

```python

def facilities(coordinates,maxdistance):
    facilities =  {
        'starbucks': ge.geoQuery(starbucks,coordinates,maxdistance,0),
        'schools' : ge.geoQuery(schools,coordinates,maxdistance,0),
        'nightclubs' : ge.geoQuery(nightclubs,coordinates,maxdistance,0),
        'vegan rest' : ge.geoQuery(vegan,coordinates,maxdistance,0),
        'kindergarten' : ge.geoQuery(kinder,coordinates,maxdistance,0)
    }
    return facilities
```

```python
f1 = facilities(coordinates,500)
```

```python
map1=folium.Map(location=[47.603364,-122.333359], zoom_start=15)
```

```python
mp.mapFolium(map1,optify,'euro','black')
mp.mapFolium(map1,f1['starbucks'],'home','darkgreen')
mp.mapFolium(map1,f1['nightclubs'],'glass','darkblue')
mp.mapFolium(map1,f1['vegan rest'],'cutlery','green')
mp.mapFolium(map1,f1['kindergarten'],'heart','purple')
mp.mapFolium(map1,f1['schools'],'home','purple')
```
